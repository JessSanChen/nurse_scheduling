```python
import pandas as pd
import json

# Load JSON data
with open('data/W1-03.json') as f:
    data = json.load(f)

# Convert 'Contracts' data to DataFrame
contracts_df = pd.json_normalize(data['Contracts'])
print("Contracts DataFrame:")
display(contracts_df)

# Convert 'Nurses' data to DataFrame
nurses_df = pd.json_normalize(data['Nurses'], 
                              sep='_', 
                              meta=['ID', 'contract_id', 'gender'], 
                              record_path=['Abilities', 'ShiftsInAbilities'])
print("\nNurses DataFrame:")
display(nurses_df)

# Convert 'Requirements' data to DataFrame
requirements_df = pd.json_normalize(data['Requirements'], 
                                    sep='_', 
                                    meta='skill', 
                                    record_path=['SkillRequirements'])
print("\nRequirements DataFrame:")
display(requirements_df)

# Convert 'Preassignments' data to DataFrame
preassignments_df = pd.DataFrame(data['Preassignments'])
print("\nPreassignments DataFrame:")
display(preassignments_df)

# Convert 'Shiftoffrequests' data to DataFrame
shift_off_requests_df = pd.DataFrame(data['Shiftoffrequests'])
print("\nShift Off Requests DataFrame:")
display(shift_off_requests_df)

# Convert 'Skills' data to DataFrame
skills_df = pd.json_normalize(data['Skills'], sep='_', record_path='Shifts', meta='id')
print("\nSkills DataFrame:")
display(skills_df)
```
